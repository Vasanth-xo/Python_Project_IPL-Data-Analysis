```python
#IPL Data Analysis 

"""To perform a comprehensive data mining analysis on IPL match records to identify 
the top-performing player overall and the top player of the 2016 season, based on individual run scores."""

import pandas as pd

Batsman_Scored = pd.read_excel(r"D:\IPL_Data\Batsman_Scored.xlsx")
Batsman_Scored.head()
```

```python
Match = pd.read_excel(r"D:\IPL_Data\Match.xlsx")
Match.head()
```

```python
Player_Match = pd.read_excel(r"D:\IPL_Data\Player_Match.xlsx")
Player_Match.head()
```

```python
Player = pd.read_excel(r"D:\IPL_Data\Player.xlsx")
Player.head()
```

```python
Ball_by_Ball = pd.read_excel(r"D:\IPL_Data\Ball_by_Ball.xlsx")
Ball_by_Ball.head()
```

```python
#Renaming Striker to Player_Id

BBB = Ball_by_Ball.rename(columns= {"Striker":"Player_Id"})
BBB.head()
```

```python
#Merging the data using key
#Merging Batsman_Scored and Match using "Match_Id" as key

Merge1 = pd.merge(Batsman_Scored,Match, on = "Match_Id", how = "inner") 
Merge1.head()
```

```python
#Merging Player_Match and Player using "Player_Id" as key

Merge2 = pd.merge(Player_Match,Player, on = "Player_Id", how = "inner")
Merge2.head()
```

```python
#Merging Merge1 and Merge2 using "Match_Id" as key 

Merge3 = pd.merge(Merge1,Merge2, on = "Match_Id", how = "inner")
Merge3.head()
```

```python
#Merging Merge3 and BBB using "Match_Id","Player_Id","Over_Id","Ball_Id","Innings_No" as key 

Ipl_Data= pd.merge(Merge3,BBB, on = ["Match_Id","Player_Id","Over_Id","Ball_Id","Innings_No"], how = "inner" )
Ipl_Data.head()
```

```python
#Slicing the year seperately from "Match_Date" using datetime function
Ipl_Data["Year"] = Ipl_Data["Match_Date"].dt.year
Ipl_Data.drop_duplicates().inplace = True
Ipl_Data.head()
```

```python
#Finding "Top Player" of 5 Seasons who has high run socres using groupby and sorting out

Top_10_player = Ipl_Data.groupby("Player_Name")["Runs_Scored"].sum().sort_values(ascending=False).reset_index()
Top_10_player.head(10)
```

```python
#Finding "Top Player" of 2016 who has high run socres using groupby and sorting out

Result_2016 = Ipl_Data[Ipl_Data["Year"]==2016]
Top_10_player_2016 = Result_2016.groupby("Player_Name")["Runs_Scored"].sum().sort_values(ascending = False).reset_index()
Top_10_player_2016.head(10)                                               
```
